```python
def cost(points,m):
    total_cost=0
    M=len(points)
    for i in range(M):
        total_cost+=(1/M)*((y[i]-np.dot(m,x[i]))**2)
    return total_cost
```

```python
def step_gradient(points,learning_rate,m):
    m_slope=np.zeros(9)
    M=len(points)
    new_m=np.zeros(9)
    for i in range(M):
        m_slope +=np.dot(x[i],(y[i]-np.dot(m,x[i])))*(-2/M)
    for j in range(9):
        new_m[j]=m[j]-learning_rate*m_slope[j]
    return new_m
```

```python
def gd(points,learning_rate,num_iteration):
        m=np.zeros(9)
        for i in range(num_iteration):
                m=step_gradient(points,learning_rate,m)
                print(i,'cost:',cost(points,m))
        return m
```
